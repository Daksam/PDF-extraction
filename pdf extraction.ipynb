```python
# importing the PyPDF2 library
import PyPDF2 as pdf
```

```python
#opening the pdf file using the open command
pdf_file = open(r"C:\Users\LENOVO\Documents\digicap\[Walter_Nicholson,_Christopher_M._Snyder]_Intermed(BookFi.org).pdf", "rb")
```

```python
#reading the file
read_pdf = pdf.PdfFileReader(pdf_file)
```

```python
#getting the number of pages in the file
number_of_pages = read_pdf.getNumPages()
```

```python
print(number_of_pages)
```

```python
# printing pages 78 to 87 content using the getpage method from for loop in range
for i in range(77, 87):
    pageObj = read_pdf.getPage(i)
    print(pageObj.extractText())
```
